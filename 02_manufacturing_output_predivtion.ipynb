```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x= torch.arange(0,17, dtype=torch.float32).unsqueeze(1)
y= 20 * x + 64
print(x)
print(y)
```

```python
model=nn.Linear(1,1)
loss_function=nn.MSELoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)

```

```python
epochs=80000
for epoch in range (epochs):
   optimizer.zero_grad()
   output=model(x)
   loss=loss_function(output,y)
   loss.backward()
   optimizer.step()
   if (epoch+1)%1000==0:
    print(f"epoch: {epoch+1},Loss:{loss.item()}")

```

```python
with torch.no_grad():
  new_value=torch.tensor([
    [15.0]
])
prediction=model(new_value)
print(f"prediction:{prediction.item()}")
```
